```python
#Menghitung Rating dari order_reviews dan Jumlah Value dari masing-masing rating
rating = list(df_reviews['review_score'])
rekap_rating = {}
for data in rating:
    if data in rekap_rating:
        rekap_rating[data] += 1
    else:
        rekap_rating[data] = 1
print("Menghitung Rating dari order_reviews dan Jumlah Value dari masing-masing rating")
display(rekap_rating)
print()

#Rating diurutkan secara Ascending Berdasarkan Keys
print("Rating diurutkan secara Ascending Berdasarkan Keys")
rekap_rating_sorted = dict(sorted(rekap_rating.items()))
display(rekap_rating_sorted)
print()

#Rating diurutkan secara Descending Berdasarkan value
print("Rating diurutkan secara Descending Berdasarkan value")
rating_sorted_value = dict(sorted(rekap_rating.items(), key=lambda x: x[1], reverse=True))
display(rating_sorted_value)
print()
```

```python
ratings = rekap_rating_sorted
#Diagram batang rekapitulasi rating dari order_reviews_dataset
ratings_descending = {5: 57328, 4: 19142, 1: 11424, 3: 8179, 2: 3151}
# Membuat DataFrame dari dictionary
df_descending = pd.DataFrame(list(ratings_descending.items()), columns=['Rating', 'Jumlah'])

# Mengurutkan DataFrame berdasarkan kolom 'Jumlah' secara descending
df_descending = df_descending.sort_values(by='Jumlah', ascending=False)
# Plotting diagram batang
df_descending.plot(kind='bar', x='Rating', y='Jumlah', legend=False)

# Menambahkan label di atas setiap batang diagram
for index, value in enumerate(df_descending['Jumlah']):
    plt.annotate(str(value), (index, value), ha='center', va='bottom', xytext=(0, 1), textcoords='offset points')
# Menambahkan judul dan label pada sumbu-sumbu & Menampilkaan diagram batang
plt.title('Rekapitulasi Review Score secara Descending')
plt.xticks(rotation=360)
plt.xlabel('Review_score')
plt.ylabel('Jumlah')
plt.show()

print()

# Plotting diagram lingkaran
plt.pie(ratings.values(),
         labels=ratings.keys(),
         autopct='%1.1f%%',
         startangle=140)
# Menambahkan label di atas setiap diagram lingkaran & Menampilkaan diagram lingkaran
plt.axis('equal')
plt.title('Presentase Review Score Secara Descending')
plt.show()

print()

# Menghitung rata-rata review score
rata_rata_rating = df_reviews['review_score'].mean()
# Menampilkan hasil dengan pembulatan ke 1 desimal
rata_rata_rating_bulat = round(rata_rata_rating,1)
print("Rata-rata Rating dari file order_reviews             :",rata_rata_rating)
print("Rata-rata Rating dari file order_reviews dibulatkan  :",rata_rata_rating_bulat)
```

```python
#Inisialisasi Variabel review score & jumlah review
review_scores = [1, 2, 3, 4, 5]
jumlah_review = [9406, 2941, 7961, 18987, 57066]
df = pd.DataFrame({'Review Score': review_scores, 'Jumlah Review': jumlah_review})
# Mengurutkan DataFrame berdasarkan kolom 'Jumlah Review' secara ascending
df.sort_values(by='Jumlah Review', inplace=True)

# Plotting diagram batang setelah pengurutan
ax = df.plot(kind='bar', x='Review Score', y='Jumlah Review', color='skyblue')
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 4), textcoords='offset points', fontsize=8, color='black', fontweight='bold')
# Menambahkan judul dan label pada sumbu-sumbu & Menampilkaan diagram batang
plt.xlabel('Review Score')
plt.xticks(rotation=360)
plt.ylabel('Jumlah Review')
plt.title('Jumlah Review untuk Order Status "Delivered" (Urutkan Ascending)')
plt.show()
print()
# Plotting diagram lingkaran
plt.pie(jumlah_review,
         labels=review_scores,
         autopct='%1.1f%%',
         startangle=140)
# Menambahkan label di atas setiap diagram lingkaran & Menampilkaan diagram lingkaran
plt.axis('equal')
plt.title('Presentase  Order Status (Delivered) Urutkan Ascending')
plt.show()

print()



# Menghitung rata-rata review score
rata_rata_review_score = np.average(review_scores, weights=jumlah_review)

# Menampilkan hasil dengan pembulatan ke 1 desimal
rata_rata_review_score_bulat = round(rata_rata_review_score, 1)

# Menampilkan hasil
print("Rata-rata Review Score untuk Order Status (Delivered) Urutkan Ascending            : ",rata_rata_review_score)
print("Rata-rata Review Score untuk Order Status (Delivered) Urutkan Ascending dibulatkan : ",rata_rata_review_score_bulat)
```
